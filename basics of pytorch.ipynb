```python
import torch
import numpy as np
```

a rank zero tensor is a scalar, \
by default its **float**


```python
# PyTorch
a = torch.tensor(1.)
print(a)
print(a.shape)
```

rank 1 tensor = vector

```python
a = torch.tensor([1., 2., 3.])
print(a)
print(a.shape)
```

rank 2 tensor = matrix

```python
a = torch.tensor([
  [1., 2., 3.],
  [2., 3., 4.]
])

print(a.shape)
print(a.ndim)#dim=num of rows
print(a.dtype)
```


### **3D Tensor**
An RGB image. It is a stack of matrices\
3D tensor, one matrix on top of another

```python
a = torch.tensor([
  [
    [1., 2., 3.],
    [2., 3., 4.]
  ],
  [
    [5., 6., 7.],
    [8., 9., 10.]
  ]
])
print(a.shape)
```


We can think of PyTorch tensors as NumPy arrays with some additional functionality, such as automatic differentiation and computations on a GPU

```python
#to convert from numpy
np_array = np.array([1., 2., 3.])
a = torch.from_numpy(np_array)
print(a)
# OR
a = torch.tensor(np_array)
print(a)
```

```python
# to change datatyoe
print(a.dtype)
a = a.to(torch.float32)
print(a.dtype)
```

```python
#give this tensor:
a = torch.tensor([
  [1., 2., 3.],
  [2., 3., 4.]
])
print(a)
#to change shape you can use this
print(a.view(3, 2))
```
